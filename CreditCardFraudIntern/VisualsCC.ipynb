```python
'''
In this notebook we will visually look at the credit card data set. Since I am a beginner to Machine Learning, 
I did a thourough look at the data set. Atleast things which I thought were important to visualize for the machine learning.
I received the data from Kaggle

> https://www.kaggle.com/mlg-ulb/creditcardfraud

Throughout the program I distinguished fraud with the color orange and non-fraud/normal with purple.
August 9, 2020
aharris26@live.esu.edu
Arlyn Harris
'''
#Import packages needed for visualization of the dataset
%matplotlib inline
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#We will import the credit card dataset
credit = pd.read_csv('C:\\Users\\arlyn\\Documents\\ESU\\CreditCardFraudIntern\\creditcard.csv')

#Peek at the data to make sure it was imported
credit.head()
```

```python
#Get information on the size and shape of the data
print("The rows and columns of the data are:", credit.shape)
```

```python
#Print out names of the data columns
credit.columns
```

```python
#Print out the data type of each column index. 
credit.info()
```

```python
#We will now check for any missing value in the dataset
credit.isnull().sum()
```

```python
#Since we understand the dataset has no missing values, we will now get some basic statistical information about it
credit.describe()
```

```python
#We will now see a visual representation of the data
credit.hist(figsize=(20,20), color='purple')
plt.show()
```

```python
#We will now distinguish the data bewteen fraud and non-fraud in a visual aid
label='Fraud', 'Non-Fraud'
info = [credit[credit.Class==1].Amount.count(), credit[credit.Class==0].Amount.count()]
fig, ax = plt.subplots(figsize=(10,10))
ax.pie(info, labels=label, colors=['orange', 'purple'])
plt.show()
```

```python
#We will now work to get a better idea of the fraud and non-fraud data
#Fraud data
fraud = credit[credit["Class"]==1]
fraud.Amount.describe()
```

```python
#Non-fraud data
nonfraud=credit[credit["Class"]==0]
nonfraud.Amount.describe()
```

```python
print("Total number of fraud transactions:", fraud.Amount.count(), " or ", len(fraud)/len(credit)*100,"%")
```

```python
print("Total number of non-fraud transactions:", nonfraud.Amount.count(), " or ", len(nonfraud)/len(credit)*100,"%")
```

```python
#Create a correlation map to better observe which columns have a relation
cm = credit.corr()
fig = plt.figure(figsize=(15, 15))
sb.heatmap(cm, cmap='PRGn')
plt.title("Correlation Matrix of Credit Card Fraud Dataset")
plt.xlabel("Credit Card Columns")
plt.ylabel("Credit Card Columns")
plt.show()
```

```python
#Take a closer look at the correlation that we saw in the heatmap
yValue = credit.Time

#Have divided the columns into 2 so that is will process a bit quicker.
interest_first_half = credit[['V1', 'V2', 'V3','V4','V5','V6','V7','V8', 'V9']]
interest_second_half=credit[['V10', 'V11', 'V12','V13','V14','V15','V16', 'V17','V18']]
```

```python
#For loop to display the scatterplots of the columns that were of interest to us.
for i in interest_first_half:
    sb.scatterplot(x=credit[i], y=yValue, hue=credit.Class, edgecolors='black', linewidth=0, palette=['purple', 'orange'])
    plt.show()
```

```python
for i in interest_second_half:
    sb.scatterplot(x=credit[i], y=yValue, hue=credit.Class, edgecolors='black', linewidth=0, palette=['purple', 'orange'])
    plt.show()
```
